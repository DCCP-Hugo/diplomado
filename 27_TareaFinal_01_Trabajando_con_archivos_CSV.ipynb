## <font color='green'>Tarea 4:</font> Extrae los emails si existen?

1. Extrae los 20 primeros emails de los registros; si es que existen, por supuesto!!<br>
2. Guárdalos en una lista y comprueba que sean 20.
3. Imprimelos desde la lista ordenadamente hacia abajo

````python
import csv
with open('/content/drive/My Drive/carpeta principal grupo 2/example.csv','r') as csv_fp:
    csv_rdr = csv.reader(csv_fp, delimiter=',')

    reg_num = 0
    first_20_email = []
    for fila in csv_rdr:              
        if reg_num == 0:
            col_email = fila.index('email')            
        elif len(first_20_email) < 20 and fila[col_email].find('@'):
            first_20_email.append(fila[col_email])
        reg_num += 1    

print(first_20_email)
print(f'Total de registros {len(first_20_email)}')
csv_fp.close()
````

````python
#para imprimir ordenadamente hacia abajo, hacemos un ciclo for que recorra la lista
for row in first_20_email:
  print(row)
````

## <font color='green'>Tarea 5:</font> Extrae los nombres reales completos

1. Extrae los 20 primeros nombres completos **de mujeres** de los registros.
2. Guárdalos en una lista y comprueba que sean 20.
3. Imprimelos desde la lista ordenadamente hacia abajo.

````python
import csv
with open('/content/drive/My Drive/carpeta principal grupo 2/example.csv','r') as csv_fp:
    csv_rdr = csv.reader(csv_fp, delimiter=',')

    reg_num = 0
    first_20_full_name_female = []
    for fila in csv_rdr:              
        if reg_num == 0:
            col_first_name = fila.index('first_name')
            col_last_name = fila.index('last_name')
            col_gender = fila.index('gender')
           
        elif len(first_20_full_name_female) < 20 and fila[col_gender].find('Female') == 0:
            first_20_full_name_female.append(fila[col_first_name]+' '+fila[col_last_name])
        reg_num += 1

print(first_20_full_name_female)
print(f'Total de registros {len(first_20_full_name_female)}')
csv_fp.close()   
````

````python
#para imprimir ordenadamente hacia abajo, hacemos un ciclo for que recorra la lista
for row in first_20_full_name_female:
  print(row)
````

## <font color='green'>Tarea 6:</font> Escribe en un archivo csv nuevo

1. Toma los nombres de la Tarea 5 y escríbelos en un archivo *cvs*. Llama al archivo *agentes_vigilar**

We can also write csv files, either new ones or add on to existing ones.

Tips:
1. Abre un archivo nuevo en modo escritura (*modo w*)
2. Escribe en el desde la lista utilizando un ciclo *for* y el método **writer** de la librería **csv**


````python
import csv
with open('/content/drive/My Drive/carpeta principal grupo 2/agentes_vigilar.csv','w', newline='') as csv_fp_agentes_vigilar:
    agentes_vigilar_wrt = csv.writer(csv_fp_agentes_vigilar)
    writer = csv.writer(csv_fp_agentes_vigilar)

    #para escribir el encabezado
    writer.writerow(['Nombre'])
    #recore cada nombre de la lista y lo escribe en el csv con el metodo writerow
    for i in range(len(first_20_full_name_female)):
      writer.writerow([first_20_full_name_female[i]])
````

Importante notar que si hacemos el writerow directamente desde un string, se toma como que cada palabra fuera una columna diferente, por lo tanto, se deben usar listas. Por eso, para escribir el encabezado, se utilizó [´Nombre'] en vez de solo 'Nombre'.
